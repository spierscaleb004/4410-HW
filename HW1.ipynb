```python
# Exercise 9.16

import pandas as pd
import matplotlib.pyplot as plt

#Step A and B
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
#Treat the first column as the index
df = pd.read_csv(url, index_col=0)

#Step C
print("First 7 rows: ")
display(df.head(7))

#Step D
print("\nLast 7 rows: ")
display(df.tail(7))

#Step E
#Reset index to treat index as a regular column
print("\nDescriptive statistics (numeric columns):")
df_reset = df.reset_index()
numeric_cols = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
display(df_reset[numeric_cols].describe())

#Step F
print("\nDescriptive statistics (categorical columns):")
categorical_cols = ['cut', 'color', 'clarity']
display(df_reset[categorical_cols].describe())

#Step G
print("\nUnique values in 'cut' column:")
for col in categorical_cols:
    print(f"{col}: {df_reset[col].unique()}")

print("\nUnique values in 'color' column:")
print(df['color'].unique())

print("\nUnique values in 'clarity' column:")
print(df['clarity'].unique())

#Step H
df_reset = df.reset_index()
numeric_cols = ['carat','depth','table','price','x','y','z']
df_reset[numeric_cols].hist(bins=20, figsize=(10, 10))
#Adjust layout so subplots do not overlap or get cut off
plt.tight_layout()
#Display
plt.show()
```
